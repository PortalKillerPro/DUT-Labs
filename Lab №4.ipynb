Порівняти значення інтерполяційних поліномів в точках:

$x=g-2k*2.7, g-2k-0.5, g-2k+2.8;$ 
 зі значеннями точної функції, оцінити точність наближення.

```python
import numpy as np

g = 3
k = 8

def f(x):
    return np.sin(x) + x
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
from scipy.interpolate import lagrange as lan

x = np.array([g - 2 * k * 2.7, g - 2 * k - 0.5, g - 2 * k + 2.8], float)
y = f(x)

xt = [g-2*k*2.7, g-2*k-0.5, g-2*k+2.8]
yt = f(xt)

plt.figure(figsize=(16, 9))
plt.style.use('ggplot')
plt.rcParams['text.color'] = "black"



poly = lan(x, y)
u = plt.plot(x, y,'bo', label = "Dots")
u = plt.plot(xt, yt,'kv', label = "Expected Dots")
t = np.linspace(0, 1, len(x))
pxLagrange = lan(t, x) 
pyLagrange = lan(t, y) 
n = 100
ts = np.linspace(t[0], t[-1], n)
xLagrange = pxLagrange(ts) 
yLagrange = pyLagrange(ts) 
plt.plot(xLagrange, yLagrange,'g',label = "Polynomial")
plt.legend(loc=4, prop={'size': 20})
plt.show()

a = [1,2,3]
print(x)
print(xLagrange)
print(y)
print(yLagrange)
```
