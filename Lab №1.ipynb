2. Побудувати графік функції: $f(x)=(x-2g*k)^2$,
де g – остання цифра у номері студентського квитка, а 
k – передостання. 
Знайти інтервали для коренів рівняння $f(x)=0$

```python
g = 3
k = 8

def f(x):
    return np.power((x - 2 * g * k), 2) - np.sin(x - 2 * g * k)
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import scipy as sp
import numpy as np
from scipy import optimize

x = np.linspace(-1e3, 1e3)

roots = optimize.fsolve(f, [0, 0])
print(roots)

plt.figure(figsize=(16, 9))
plt.style.use('ggplot')
plt.rcParams['text.color'] = "black"
plt.plot(x, np.zeros(len(x)))
plt.plot(x, f(x), 'g', label="F(x)")
plt.plot(roots, f(roots), 'kv', label="root")
plt.legend(loc="best", prop={'size': 20})
plt.show()
```

3. Знайти корінь трансцендентного рівняння: $f (x) = 0$

```python
print(optimize.newton(f, 0))
```
