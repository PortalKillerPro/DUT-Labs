Для функції $f(x)=х3+2x$ знайти похідну за допомогою методів послідовності відношень приростів та центрованої формули другого порядку
у точці $х=g–2k–2$ з точністю $0.5$. 

Порівняти із значенням аналітичної похідної функції, оцінити точність по кожному методу.

```python
import numpy as np

g = 4
k = 8

def f(x):
    return x * 3 + 2 * x
```

```python
from scipy.misc import derivative

x = g - 2 * k
y = np.linspace(-434.25, 434.25)
dx = derivative(f, x, dx=0.5)
dy = derivative(f, y, dx=0.5)
print(dx)
```

На основі складеної таблиці за допомогою формули квадратичної
інтерполяції знайти значення похідної в точці $х=g–2k–2$. 

Порівняти із значенням аналітичної похідної функції, оцінити точність.

```python
%matplotlib inline

from scipy.interpolate import UnivariateSpline as US
import numpy as np
import matplotlib.pyplot as plt

x = np.array([g-2*k-3, g-2*k-1, g-2*k, g-2*k+1], dtype="float")
y = f(x)
spl1 = US(x, y, s=1e-5)
dydx = spl1.derivatives(g-2*k-2)
print(x,y)
print(dydx[1])

plt.figure(figsize=(16, 9))
plt.style.use('ggplot')
plt.rcParams['text.color'] = "black"

plt.plot(x,y,'-x', label="f(x)")
plt.plot(x,y,'*', label="def f(x)")
plt.legend(loc=4, prop={'size': 20})
plt.show()
```
